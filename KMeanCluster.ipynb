```python
import pandas as pd
```

```python
df = pd.read_csv('income.csv')
```

```python
df
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df['Age'],df['Income($)'],color = 'blue',marker = '*')
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3)
km
```

```python
y_pred = km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
df['cluster'] = y_pred
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```

```python
plt.scatter(df1['Age'],df1['Income($)'],color = 'blue',marker = '*')
plt.scatter(df2['Age'],df2['Income($)'],color = 'green',marker = '*')
plt.scatter(df3['Age'],df3['Income($)'],color = 'red',marker = '*')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])                                
df
```

```python
km = KMeans(n_clusters = 3)
km
```

```python
y_pred1 = km.fit_predict(df[['Age','Income($)']])
y_pred1
```

```python
df['cluster'] = y_pred1
```

```python
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1['Age'],df1['Income($)'],color = 'blue',marker = '*')
plt.scatter(df2['Age'],df2['Income($)'],color = 'green',marker = '*')
plt.scatter(df3['Age'],df3['Income($)'],color = 'red',marker = '*')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters = k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel('K')
plt.ylabel('sum of square error')
plt.plot(k_rng,sse)
```
